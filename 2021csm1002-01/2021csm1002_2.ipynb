```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
import math


def generate_data_points(size):
    np.random.seed(42)
    noise = np.random.normal(0,0.004,size)

    ran = np.random.uniform(0,360,size)
    # print(ran)

    X=[]
    Y=[]
    for i in range(size):
        x = ran[i]
        y = math.cos(2*np.pi*(ran[i]*np.pi/180))+noise[i]+ran[i]/360
        X.append(x)
        Y.append(y)

    return X,Y
    

```

```python
def polynomial_regression(n,X,Y):
    
    A = []
    b = []

    for i in range(n):
        # print("row : ",i)
        A_r = []
        for j in range(n):
            # print("column : ",j)
            # print("element : ", f'sum(x**{i+j})')
            A_r.append(np.sum([x**(i+j) for x in X]))
        A.append(A_r)
        b.append(np.sum(np.multiply([x**i for x in X], Y)))
    w = np.linalg.inv(A) * np.transpose(b)
    return w
        
    # print("A = ",A)
    # print("b = ",b)
    
def predict(X,w): 
    pred = [] 
    for x in X:
        pred.append(np.sum([w_j * x**j for j, w_j in
                enumerate(w)])) 
    return pred

def polynomial_regression_for_multiple_degrees(size):
    X,Y = generate_data_points(size)
    degree_power = [1,2,3,5,7,10]
    for d in degree_power:
        w = polynomial_regression(d,X,Y)
        plt.scatter(X, Y)
        x_min = min(X)
        x_max = max(X)
        xx = np.arange(x_min, x_max, 0.05)
        plt.plot(xx, predict(xx,w), c="red")
        print('degree_power = ',d)
        print('data_size : ',size)
        plt.show()
```

```python
size = 20

for size in range(20,120,20):
    polynomial_regression_for_multiple_degrees(size)
```
